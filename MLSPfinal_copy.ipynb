#zero crossing rate:

```python
import librosa.display
x , sr = librosa.load(data_path+'/rock/rock.00099.wav')
#Plot the signal:
plt.figure(figsize=(14, 5))
librosa.display.waveplot(x, sr=sr)
zero_crossings = librosa.zero_crossings(x, pad=False)
print(sum(zero_crossings))
```

```python
sum(zero_crossings)/x.shape[0] #zero_crossing_rate_mean
np.var(librosa.feature.zero_crossing_rate(x)[0]) #zero_crossing_rate_var
```
